## Redes de Carreteras con Python

# Obtenci√≥n de datos de la red de carreteras OpenStretMap (OSM)
import osmnx as ox  

```python
admin = ox.geocode_to_gdf('Asuncion')
admin_poly = admin.geometry.values[0]
admin_poly
```

```python
G = ox.graph_from_polygon(admin_poly)
```

```python
G.number_of_nodes(), G.number_of_edges()
```

```python
nodes,edges = ox.graph_to_gdfs(G)
```

```python
import matplotlib.pyplot as plt 
plt.style.use('dark_background')
```

```python
f, ax = plt.subplots(1,1,figsize = (10,10))

edges.plot(ax=ax, edgecolor = 'orange', linewidth = 0.3)
ax.axis('off')
```
